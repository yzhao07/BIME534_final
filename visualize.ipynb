```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

features = np.load("features.npy")         
labels = np.load("labels.npy")
pca = PCA(n_components=10)
features_pca = pca.fit_transform(features)
```

```python
unique_labels = np.unique(labels)
colors = plt.cm.tab10(np.linspace(0, 1, len(unique_labels)))  # color palette
label_to_color = {label: colors[i] for i, label in enumerate(unique_labels)}

# Step 3: Plot
plt.figure(figsize=(8, 6))
for label in unique_labels:
    idx = labels == label
    plt.scatter(features_pca[idx, 0], features_pca[idx, 1],
                color=label_to_color[label], label=label, s=10)

plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.title("PCA of CONCH Patch Features")
plt.legend(markerscale=2, fontsize='small')
plt.grid(True)
plt.tight_layout()
plt.show()
```
